## Function Definitions


```python
import unicodedata
import re
def normalize_text(text: str) -> str:
    if not isinstance(text, str):
        return ""
    text = text.lower()
    # remove diacritics (ġ → g, ż → z, għ → gh, ħ → h, ecc.)
    text = unicodedata.normalize("NFKD", text)
    text = text.encode("ascii", "ignore").decode("ascii")
    # keep only letters and spaces
    text = re.sub(r"[^a-z]+", " ", text)
    # collapse multiple spaces
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
# Build variant_to_canonical mapping for town extraction
# This maps normalized variant strings -> canonical town name
variant_to_canonical = {}
for canonical, variants in locality_variants.items():
    # include canonical name normalized
    variant_to_canonical[normalize_text(canonical)] = canonical
    for v in variants:
        variant_to_canonical[normalize_text(v)] = canonical

# Sanity check
print(f"Built variant_to_canonical with {len(variant_to_canonical)} entries")
```

```python
def extract_town_from_text(text: str):
    """Return the first canonical town whose variant appears in `text` and
    is followed by a full stop (period) in the original text.
    """
    if not isinstance(text, str):
        return None

    # Normalized form for variant matching (letters + spaces)
    norm_text = normalize_text(text)
    padded = f" {norm_text} "  # helps avoid partial word matches

    # Preserve periods in a normalized string for checking trailing full stop
    preserve = unicodedata.normalize("NFKD", text)
    preserve = preserve.encode("ascii", "ignore").decode("ascii").lower()
    # keep letters and periods only (replace others with spaces)
    preserve = re.sub(r"[^a-z\.]+", " ", preserve)
    padded_preserve = f" {preserve} "

    # Find the first town mentioned by text position, but require a trailing full stop
    first_town = None
    first_position = len(padded)

    for variant_norm, canonical in variant_to_canonical.items():
        pattern = f" {variant_norm} "
        position = padded.find(pattern)
        if position != -1 and position < first_position:
            # Require that in the original (preserve) text the variant is followed by a full stop.
            # Allow optional whitespace between the variant and the dot.
            if re.search(rf" {re.escape(variant_norm)}\s*\.", padded_preserve):
                first_position = position
                first_town = canonical

    return first_town
```

## API Function Definitions

```python
import requests
from datetime import datetime, timedelta
import time

def get_weather_for_location(lat, lon, date_str=None):
    """
    Fetch weather data from Open-Meteo API (free, no API key required)
    
    Args:
        lat: Latitude
        lon: Longitude
        date_str: Date string in format 'YYYY-MM-DD' (if None, gets current weather)
    
    Returns:
        Dictionary with weather data or None if request fails
    """
    try:
        if date_str:
            # Historical weather - use archive API
            url = "https://archive-api.open-meteo.com/v1/archive"
            params = {
                "latitude": lat,
                "longitude": lon,
                "start_date": date_str,
                "end_date": date_str,
                "daily": "temperature_2m_max,temperature_2m_min,temperature_2m_mean,precipitation_sum,rain_sum,windspeed_10m_max",
                "timezone": "Europe/Malta"
            }
        else:
            # Current weather
            url = "https://api.open-meteo.com/v1/forecast"
            params = {
                "latitude": lat,
                "longitude": lon,
                "current": "temperature_2m,precipitation,rain,windspeed_10m",
                "timezone": "Europe/Malta"
            }
        
        response = requests.get(url, params=params, timeout=10)
        response.raise_for_status()
        return response.json()
    except Exception as e:
        print(f"Error fetching weather for ({lat}, {lon}): {e}")
        return None

# Test with Valletta coordinates
test_weather = get_weather_for_location(35.8989, 14.5136)
if test_weather:
    print("Successfully connected to weather API!")
    print(f"Sample response structure: {list(test_weather.keys())}")
else:
    print("Failed to connect to weather API")
```

## Extracting Street Names from Policy Articles

```python
# Extract street names from police releases
# Common patterns: "Triq [name]", "Street", "Road", "Avenue", etc.
import re

street_patterns = [
    r"\b(Triq\s+(?:[A-ZĦŻĠa-zħżġ'\-]+\s+){1,5}[A-ZĦŻĠa-zħżġ'\-]{3,})",  # "Triq [name]" - last word must be at least 3 chars
    r"(?:in|on|along|at)\s+([A-ZĦŻĠa-zħżġ'\-]+(?:\s+[A-ZĦŻĠa-zħżġ'\-]+){0,3}\s+(?:Street|Road|Avenue|Square|Wharf|Drive|Bypass))\b",  # English street names
]

streets_found = set()

for text in police_releases_df["content"].dropna().tolist():
    for pattern in street_patterns:
        matches = re.findall(pattern, text)
        for match in matches:
            cleaned = match.strip()
            # Filter out obvious non-streets and incomplete names
            words = cleaned.split()
            last_word = words[-1] if words else ""
            
            if (len(words) >= 2 and 
                len(words) <= 6 and
                len(last_word) >= 3 and  # Last word must be at least 3 characters
                not any(bad in cleaned.lower() for bad in ['mater dei', 'hospital', 'following', 'accident', 'traffic', 'general hospital'])):
                streets_found.add(cleaned)

# Sort and display first 50 street names
sorted(list(streets_found))[:50]
```

```python
# Extract street names from news articles as well
streets_news = set()

for text in articles_df["content"].dropna().tolist():
    for pattern in street_patterns:
        matches = re.findall(pattern, text)
        for match in matches:
            streets_news.add(match.strip())

# Sort and display first 50 street names from news
sorted(list(streets_news))[:50]
```

```python
# Build reverse lookup dictionary
street_to_canonical = {}
for canonical, variants in street_variants.items():
    for variant in variants:
        street_to_canonical[variant] = canonical

def extract_street_from_text(text):
    """
    Extract street name from text using the street_variants dictionary.
    Returns the canonical street name if found, otherwise None.
    """
    if pd.isna(text):
        return None
    
    text_normalized = normalize_text(text)
    
    # Check for each variant
    for variant in street_to_canonical.keys():
        # Use word boundaries to match whole words/phrases
        pattern = r'\b' + re.escape(variant) + r'\b'
        if re.search(pattern, text_normalized, re.IGNORECASE):
            return street_to_canonical[variant]
    
    return None
```

```python
# Extract street names from police releases and add as new column
police_releases_df["street_clean"] = police_releases_df["content"].apply(extract_street_from_text)

# Show summary statistics
print(f"Total police releases: {len(police_releases_df)}")
print(f"Releases with identified streets: {police_releases_df['street_clean'].notna().sum()}")
print(f"Releases without identified streets: {police_releases_df['street_clean'].isna().sum()}")
print(f"\nUnique streets identified: {police_releases_df['street_clean'].nunique()}")

# Show the most common streets
print("\nTop 20 most frequently mentioned streets:")
police_releases_df['street_clean'].value_counts().head(20)
```

## Weather Data for Accident Locations
Let's fetch weather data for the towns where accidents occurred to analyze potential correlations between weather conditions and accident frequency.

```python
# Rain status (boolean) for each town on last Sunday
from datetime import datetime, timedelta
import pandas as pd

# Compute last Sunday (always previous Sunday, not today if today is Sunday)
today = datetime.now().date()
delta = (today.weekday() + 1) % 7
if delta == 0:
    delta = 7
last_sunday = today - timedelta(days=delta)
last_sunday_str = last_sunday.strftime('%Y-%m-%d')

print(f"Checking rain on last Sunday: {last_sunday_str}")

# Collect towns present in the dataset
unique_towns = (
    police_releases_df['town_clean']
    .dropna()
    .astype(str)
    .str.lower()
    .unique()
)

results = []
skipped = []
for town in sorted(unique_towns):
    coords = malta_town_coords.get(town)
    if not coords:
        skipped.append(town)
        continue
    lat, lon = coords
    data = get_weather_for_location(lat, lon, date_str=last_sunday_str)
    rain_amount = 0.0
    if data and 'daily' in data:
        daily = data['daily']
        # Prefer rain_sum; fallback to precipitation_sum
        if isinstance(daily.get('rain_sum'), list) and daily['rain_sum']:
            rain_amount = daily['rain_sum'][0] or 0.0
        elif isinstance(daily.get('precipitation_sum'), list) and daily['precipitation_sum']:
            rain_amount = daily['precipitation_sum'][0] or 0.0
    results.append({
        'town': town,
        'date': last_sunday_str,
        'raining': bool(rain_amount and rain_amount > 0)
    })

rain_df = pd.DataFrame(results).sort_values('town').reset_index(drop=True)
print("\nRain status by town (True/False):")
print(rain_df)

if skipped:
    print("\nTowns skipped (no coordinates available):", ", ".join(skipped))
```
